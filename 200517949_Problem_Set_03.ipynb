```python

#Question-01

#Step 1. Import the necessary libraries

import pandas as pd

#Step 2. Import the dataset from this address (https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).
#Step 3. Assign it to a variable called users
# using to Delimiter to read the data

users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|') 

#Step 4. Discover what is the mean age per occupation

mean_age_per_occupation = users[['occupation', 'age']].groupby('occupation').mean()

#displaying the mean age per occupation
mean_age_per_occupation 
```

```python
#Step 5. Discover the Male ratio per occupation and sort it from the most to the least

df = users[['gender', 'occupation']]

# Filter Male in gender
males = df.loc[df['gender'] == 'M']

# Total Count
group_by_occupations = df.groupby('occupation').count()

# Male Count
male_count_by_occupation = males.groupby('occupation').count()

# Discover the male_ratio_per_occupation
male_ratio_per_occupation = male_count_by_occupation / group_by_occupations

male_ratio_per_occupation.sort_values(ascending=False, by='gender')



```

```python
#Step 6. For each occupation, calculate the minimum and maximum ages

df = users[['occupation', 'age']]

# Minimum Age
occupations_min_age = df.groupby('occupation').min()

#Maximum Age
occupations_max_age = df.groupby('occupation').max()

# concat both max and min
max_min_age = pd.concat([occupations_max_age, occupations_min_age], axis=1)

max_min_age


```

```python
#Step 7. For each combination of occupation and sex, calculate the mean age

df = users[['occupation', 'gender', 'age']]

users_mean_age_occupation_and_gender = df.groupby(['occupation', 'gender']).mean('age')

users_mean_age_occupation_and_gender

```

```python

#Step 8. For each occupation present the percentage of women and men

#the count in percentage
gender_percentage_by_occupation = df.groupby('occupation')['gender'].value_counts(normalize=True)*100

gender_percentage_by_occupation
```

```python
#Question -04
#Wind Statistics

#Step 1. Import the necessary libraries

import pandas as pd

#Step 2. Import the dataset from the attached file wind.txt

data = pd.read_csv('wind.txt', delimiter='\s+')

#Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

data.rename(columns={"Yr":"Year", "Mo":"Month", "Dy":"Day"}, inplace=True)

data
```

```python
#Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

def fixYear():
    data['Year'] += 1900 

fixYear()

data

```

```python
#Step 5. Set the right dates as the index.

data.index = pd.to_datetime(pd.concat([data['Year'], data['Month'], data['Day']], axis=1)) 

data.drop(['Year','Month','Day'], axis=1, inplace=True) 

data
```

```python
#Step 5. Pay attention at the data type, it should be datetime64[ns].

data.index
```

```python

#Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.

data.isna().sum()
```

```python
#sum of null values

data.isnull().sum().sum()
```

```python
#Step 7. Compute how many non-missing values there are in total.

data.notnull().sum().sum()
```

```python
#Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times. A single number for the entire dataset.

data.mean().mean()
```

```python
#Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days
#A different set of numbers for each location.

loc_stats = pd.concat([data.min(), data.max(), data.mean(), data.std()], axis=1) 

loc_stats.rename(columns={0:'Min', 1:'Max', 2:'Mean', 3:'Std'}, inplace=True) 

loc_stats
```

```python
#Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.
#A different set of numbers for each day.

day_stats = pd.concat([data.min(axis=1), data.max(axis=1), data.mean(axis=1), data.std(axis=1)], axis=1)

day_stats.rename(columns={0:'Min',1:'Max',2:'Mean',3:'Std'}, inplace=True)

day_stats

```

```python
#Step 11. Find the average windspeed in January for each location.

avg_windspeed_in_jan = data[data.index.month == 1].mean()

avg_windspeed_in_jan
```

```python
#Step 12. Downsample the record to a yearly frequency for each location.

yearly = data.resample('Y').ffill()

yearly
```

```python
#Step 13. Downsample the record to a monthly frequency for each location.

monthly = data.resample('M').ffill()

monthly
```

```python
#Step 14. Downsample the record to a weekly frequency for each location.

weekly = data.resample('W').ffill()

weekly
```

```python
#Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

weekly_stats = weekly[:52]

weekly_stats.shape[0]
```

```python
weekly_stats = pd.concat([weekly_stats.mean(), weekly_stats.max(), weekly_stats.mean(), weekly_stats.std()], axis=1)

weekly_stats.rename(columns={0:'Mean',1:'Max',2:'Mean',3:'Std'},inplace=True)

weekly_stats
```

```python
#Question 6
#Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.
#Don't forget to label your axes!

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('usa.csv')

df
```

```python
fig = plt.figure() #empty figure with no axes

plt.rcParams["figure.figsize"] = [15, 10] #figure size

df.plot( 'Year', ['Marriages_per_1000','Divorces_per_1000'], label=['Marriages', 'Divorces'])

plt.title('Marriages and Divorces per capita in the U.S. between 1867 and 2014', color = 'y', fontsize = 24)

#axis labels
plt.xlabel('Year', color = 'y', fontsize = 20)
plt.ylabel('Capita', color = 'y', fontsize = 20)

plt.show()
```

```python
#Question 7
#Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.
#Don't forget to label your axes!

filter_by_years = (df['Year'] == 1900) | (df['Year'] == 1950) | (df['Year'] == 2000) 

df_1 = df.loc[filter_by_years, ['Year','Marriages_per_1000','Divorces_per_1000']]

df_1
```

```python
df_1[{'Marriages_per_1000','Divorces_per_1000'}].plot.bar()

plt.title("Marriages and Divorces per capita in the U.S. between 1900, 1950 and 2000", color = 'y', fontsize = 24)

plt.xlabel('Year', color = 'y', fontsize = 20)
plt.ylabel('Capita', color = 'y', fontsize = 20)

plt.xticks((0, 1, 2), ("1900", "1950", "2000"), rotation = 0)
```

```python
#Question 8
#Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name.
#Don't forget to label your axes!


df = pd.read_csv('actor_kill_counts.csv')

sorted_df = df.sort_values("Count", ascending = True)

sorted_df.plot.barh(x='Actor', y='Count')

plt.title('Hollywood Actors\' Kill Count',color = 'k', fontsize=24)

plt.xlabel('Kills Count', color = 'y', fontsize = 20)
plt.ylabel('Actor', color = 'y', fontsize = 20)
```

```python
#Question 9
#Create a pie chart showing the fraction of all Roman Emperors that were assassinated.
#Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories

df = pd.read_csv('roman-emperor-reigns.csv')

death_cause = df.groupby("Cause_of_Death")['Emperor'].count()

death_cause.plot.pie(autopct="%.1f%%", fontsize = 18, ylabel = '', explode=[0.01]*8)

plt.title("Roman Emperors' Cause of Death", color = 'k', fontsize = 24, fontweight ='bold')
```
